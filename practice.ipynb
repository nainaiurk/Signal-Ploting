```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,1,1000)
c = 2+ t*0
plt.title('DC Signal')
plt.plot(t,c)
plt.xlabel('Time')
plt.ylabel('Current')
plt.show()
```

```python


t = np.linspace(0,10,1000)
c = np.exp(t)
plt.title('Exponential Signal')
plt.plot(t,c)
plt.xlabel('Time')
plt.ylabel('Current')
plt.show()
```

```python
from scipy import signal

t = np.linspace(0,1,2000)
c = signal.square(2*np.pi * 5 * t)
plt.ylim(-1.5, 1.5)
plt.grid(True)
plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
plt.title('Square Signal')
plt.xlabel('Time')
plt.ylabel('Current')
plt.plot(t,c)
plt.show()
```

```python
t = np.linspace(0,1,2000)
c = signal.sawtooth(2*np.pi * 4 * t,0.5)
plt.grid(True)
plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
plt.ylim(-1.5, 1.5)
plt.title('Triangle Signal')
plt.xlabel('Time')
plt.ylabel('Current')
plt.plot(t,c)
plt.show()
```

```python
from scipy import signal
t = np.linspace(0,1,100)
c = signal.unit_impulse(100)
plt.grid(True)
plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
plt.ylim(-1.5, 1.5)
plt.title('Unit Impulse Signal')
plt.xlabel('Time')
plt.ylabel('Current')
plt.plot(t,c)
plt.show()
```

```python
t = np.linspace(0,1,2000)
c = 1+t*0
plt.grid(True)
plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
plt.ylim(-1.5, 1.5)
plt.title('Unit Step Continuous Signal')
plt.xlabel('Time')
plt.ylabel('Current')
plt.plot(t,c,"r")
plt.show()
```
