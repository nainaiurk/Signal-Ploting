```python
import numpy as np
import matplotlib.pyplot as plt
# dc signal
t = np.linspace(0,10,100)
c = 1+t*0
plt.grid(True)
plt.ylim(-1.5,1.5)
plt.axhline(y=0,color="grey")
plt.axvline(x=0,color="grey")
plt.title('DC Signal')
plt.xlabel('Time')
plt.ylabel('Current')
plt.plot(t,c,'r')
plt.show()
```

```python
# Exponential
t = np.linspace(0,10,100)
c = np.exp(t)
plt.grid(True)
plt.axhline(y=0,color="grey")
plt.axvline(x=0,color="grey")
plt.title('Exponential Signal')
plt.xlabel('Time')
plt.ylabel('Current')
plt.plot(t,c,'r')
plt.show()
```

```python
from scipy import signal
# square signal
t = np.linspace(0,1,1000)
c = signal.square(2*np.pi*5*t)
plt.grid(True)
plt.ylim(-1.5,1.5)
plt.axhline(y=0,color="grey")
plt.axvline(x=0,color="grey")
plt.title('Square Signal')
plt.plot(t,c,'r')
plt.xlabel('Time')
plt.ylabel('Current')
plt.show()
```

```python
# Triangular Signal
t = np.linspace(0,1,1000)
c = signal.sawtooth(2*np.pi*5*t,0.5)
plt.grid(True)
plt.ylim(-1.5,1.5)
plt.axhline(y=0,color="grey")
plt.axvline(x=0,color="grey")
plt.title('Triangular Signal')
plt.plot(t,c,'r')
plt.xlabel('Time')
plt.ylabel('Current')
plt.show()
```

```python
# Unit impulse signal
t = np.linspace(0,1,100)
c = signal.unit_impulse(100,10)
plt.grid(True)
plt.ylim(-1.5,1.5)
plt.axhline(y=0,color="grey")
plt.axvline(x=0,color="grey")
plt.title('Unit Impulse Signal')
plt.plot(t,c,'r')
plt.xlabel('Time')
plt.ylabel('Current')
plt.show()
```
